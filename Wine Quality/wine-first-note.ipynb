```python
wine_label = df[['quality']]
wine_data = df.drop(['quality'],axis=1)
```

```python
wine_data.head()
```

```python
wine_data = wine_data.drop(['Id'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
lin_clf = LinearSVC(dual=True, random_state=42)
lin_clf.fit(wine_data, wine_label)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(lin_clf, wine_data, wine_label).mean()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
lin_clf = make_pipeline(StandardScaler(),
                       LinearSVC(dual=True, random_state=42))
```

```python
lin_clf.fit(wine_data, wine_label)
```

```python
lin_clf = make_pipeline(StandardScaler(),
                       LinearSVC(max_iter=1_000_000,dual=True, random_state=42))
```

```python
lin_clf.fit(wine_data, wine_label)
```

```python
cross_val_score(lin_clf, wine_data, wine_label).mean()
```

```python
from sklearn.svm import SVC
```

```python
svm_clf = make_pipeline(StandardScaler(),SVC(random_state=42))
cross_val_score(svm_clf, wine_data, wine_label).mean()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform, uniform

param_distrib = {
    'svc__gamma': loguniform(0.001, 0.1),
    'svc__C': uniform(1,10)
}
rnd_search_cv = RandomizedSearchCV(svm_clf, param_distrib, n_iter=100, cv=5, random_state=42)
```

```python
rnd_search_cv.fit(wine_data, wine_label)
```

```python
rnd_search_cv.best_estimator_
```

```python
rnd_search_cv.best_score_
```

```python
poly_kernel_svm_clf = make_pipeline(StandardScaler(),
                                   SVC(kernel='poly', degree=3, coef0=1,C=5))
cross_val_score(poly_kernel_svm_clf, wine_data, wine_label)
```

```python
rbf_kernel_svm_clf = make_pipeline(StandardScaler(),
                                  SVC(kernel='rbf', gamma=10, C=1))
cross_val_score(rbf_kernel_svm_clf, wine_data, wine_label)
```

```python
cross_val_score(rbf_kernel_svm_clf, wine_data, wine_label)
```
